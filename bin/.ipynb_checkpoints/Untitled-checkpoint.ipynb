```python
import pandas as pd
import subprocess
```

```python
# Replace 'input.gff' and 'output.gtf' with your actual file paths
input_gff = '../dl_data/a_fumigatus_Af293.gff'
output_gtf = '../data/a_fumigatus_Af293.gtf'

# Build the command to call agat_convert_sp_gxf2gxf
# agat_convert_sp_gff2gtf.pl --gff 1_test.gff -o 1_test_agat.gtf
agat_command = [
    'agat_convert_sp_gff2gtf.pl',
    '--gff', input_gff,
    '--out', output_gtf
]

# Run agat_convert_sp_gxf2gxf.py as a subprocess
subprocess.run(agat_command, check=True)
```
